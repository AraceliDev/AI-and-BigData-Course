# Reflexió: Ètica en la intel·ligència artificial.
## ¿Fins a quin punt és ètic confiar a la intel·ligència artificial decisions importants en sectors crítics com la sanitat, la justícia o l’educació?

Sense dubte, la IA té moltes avantatges significatives amb relació a la presa de decisions: la seva capacitat de processar un alt volum de dades a gran velocitat, l'emmagatzematge  i recuperació d´aquestes dades i la detecció de patrons en elles superen la capacitat cognitiva de l´ésser humà. No obstant això, en moltes ocasions no es pot comprendre les decisions d´alguns algoritmes, com els algoritmes de l´aprenentatge profund, donen lloc a la "caixa negra" on només podem conèixer l´input i l´output però no el procés de generació. És en aquest punt on no es pot assegurar que la IA respecti les normes morals establertes per la nostra societat. Per altra banda, l´aprenentatge supervisat pot oferir major transparència en les seves decisions gràcies al fet que prèviament el model s´entrena amb dades etiquetades, però això tampoc garanteix una major ètica perquè també s´hi poden donar biaxos humans. Doncs, ¿és una bona decisió l´ús de la IA en sectors crítics? Analitzarem tres escenaris per veure si arribem a una conclusió.
###  La priorització de pacients en tractaments mèdics
Uns dels algoritmes que millor s´adapta a aquest escenari és el classificador bayesià. El *teorema de Bayes* es basa en la probabilitat de A donat B, per exemple, la probabilitat de supervivència (A) donat l´historial mèdic d´un pacient, símptomes, proves, etc (B). Un model basat en aquest algoritme en un context d´urgències pot ser determinant, ja que és eficient i ràpid en el processament de múltiples variables clíniques.
No obstant això, aquest algoritme presenta limitacions, pel fet que assumeix les variables com independents, molt problemàtic si tenim en compte que en el sector sanitari els símptomes moltes vegades es troben correlacionats. Això pot donar un fals negatiu i no tractar un pacient, provocant conseqüències greus. Finalment, es pot donar priorització mèdica a grups socioeconòmics amb millor accés a atenció mèdica ampliant la bretxa sanitària.
### L’avaluació de risc de reincidència en llibertat condicional
Un model d´aprenentatge supervisat amb tècniques de classificació és interessant dins aquest àmbit. Per exemple, un model entrenat amb un algoritme de regressió logística pot donar una resposta binària, és a dir, concedir o no la llibertat condicional a una persona privada de llibertat. Donades unes variables d´entrada com, edat, nombre de delictes, tipus de delicte, nivell educatiu, consum de drogues... podria determina el risc de reincidència. Aquest algoritme interpreta bé les variables quantitatives (edat i nombre de delites) i les relacions entre les variables, però no tracta bé les variables categoriques (tipus de delicte). A més, és un model que prediu probabilitats i redueix decisions ètiques complexes a càlculs matemàtics.  
Segons l´auditoria de l´algoritmes RisCanvi(Memòria, 2024), un cas paradigmàtic és l'algoritme COMPAS (Correctional Offender Management Profiling for Alternative Sanctions), utilitzat per avaluar risc de reincidència. Investigacions de ProPublica van demostrar que aquest sistema presentava biaixos racials sistemàtics, amb taxes de falsos positius significativament més altes per a persones afroamericanes.
### L’assignació de beques o recursos educatius
En relació a l´assignació de beques o recursos, un model basat en aprenentatge supervisat amb algoritmes com arbres de decisió pot ser una alternativa interessant. És fàcil d´entendre i transparent, es pot veure per què un estudiant va rebre o no una ajuda, i pot gestionar dades numèriques i categòriques. Té com a inconvenient que és fàcil sobre ajustar el model (overfittting), creant regles massa específiques, dificultant la mobilitat social ascendent, un perfil acadèmic no convencional, però potencial pot quedar exclòs de les prestacions socials. També que canvis en les dades podem canviar completament l´arbre.

Com a conclusió, hem observat que la IA és una eina de gran ajuda en aquests escenaris, no obstant això, no podem passar per alt les seves limitacions. Es tracta de crear un model responsable amb una supervisió humana qualificada i augmentar la transparència dels algoritmes en la presa de decisions crítiques. A més, és imprescindible desenvolupar un marc legislatiu específic per regular els biaixos, establir auditories periòdiques i garantir mecanismes de control ètic.
Mentre aquest marc regulador no estigui establert, és fonamental fer un ús responsable de la IA per part de tothom, utilitzant-la com a eina de suport i recomanació, però mai com a única responsable de decisions finals. D´aquesta manera, podem aprofitar els beneficis de la seva eficiència sense comprometre els valors ètics de la nostra societat.
